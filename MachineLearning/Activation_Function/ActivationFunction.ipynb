### Activation Function 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Step Function

```python
def step_function(x):
    return np.where(x < 0, 0, 1)
```

```python
x = np.arange(-3, 3, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.show()
```

### Sigmoid Function

```python
def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-10, 10, 0.1)
y = sigmoid_function(x)

plt.plot(x, y)
plt.show()
```

```python
def sigmoid_derivative_function(x):
    return sigmoid_function(x) * (1 - sigmoid_function(x))
```

```python
derivative_y = sigmoid_derivative_function(x)

plt.plot(x, derivative_y)
plt.show()
```

### Hyperbolic Tangent Function

```python
def hyperbolic_tangent_function(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.arange(-10, 10, 0.1)
y = hyperbolic_tangent_function(x)

plt.plot(x, y)
plt.show()
```

```python
def hyperbolic_tangent_derivative_function(x):
    return 1 - hyperbolic_tangent_function(x)*hyperbolic_tangent_function(x)
```

```python
derivative_y = hyperbolic_tangent_derivative_function(x)

plt.plot(x, derivative_y)
plt.show()
```

### ReLU Function

```python
def ReLU_function(x):
    return np.where(x > 0, x, 0)
```

```python
x = np.arange(-10, 10, 0.1)
y = ReLU_function(x)

plt.plot(x, y)
plt.show()
```

```python
def ReLU_derivative_function(x):
    return np.where(x > 0, 1, 0)
```

```python
derivative_y = ReLU_derivative_function(x)

plt.plot(x, derivative_y)
plt.show()
```
